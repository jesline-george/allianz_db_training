```sql
%python
data=[(1,'Virat'),(2,'Rohit')]
schema="id int, name string"
df=spark.createDataFrame(data,schema)
df.createOrReplaceTempView("source")
```

```sql
select * from source
```

```sql
create table dev.jesline.cricket (id int, name string)
```

```sql
%python
df.write.mode("overwrite").saveAsTable("dev.jesline.cricket")
```

```sql
insert into dev.jesline.cricket select * from source
```

```sql
select * from dev.jesline.cricket 
```
